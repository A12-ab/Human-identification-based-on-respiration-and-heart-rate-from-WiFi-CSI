```python
# Separate signal and labels
heart_signals = df_hr.iloc[:, :-1].values  # All but last column
heart_labels = df_hr.iloc[:, -1].values    # Last column

resp_signals = df_rr.iloc[:, :-1].values
resp_labels = df_rr.iloc[:, -1].values
```

```python
resp_signals.shape
```

```python
import numpy as np
```

```python
def extract_time_features(signal):
    return [
        np.mean(signal),
        np.std(signal),
        np.min(signal),
        np.max(signal),
        np.ptp(signal),                # Peak-to-peak (max - min)
        np.sum(np.square(signal)),     # Energy
        np.median(signal),
        np.percentile(signal, 25),
        np.percentile(signal, 75)
    ]
```

```python
def extract_freq_features(signal, fs=100):
    fft_vals = np.abs(np.fft.rfft(signal))
    fft_freqs = np.fft.rfftfreq(len(signal), d=1/fs)
    dominant_freq = fft_freqs[np.argmax(fft_vals)]
    spectral_entropy = -np.sum((fft_vals / np.sum(fft_vals)) * np.log2(fft_vals / np.sum(fft_vals) + 1e-10))
    return [dominant_freq, spectral_entropy]
```

```python
heart_features = []
resp_features = []
```

```python
for h_sig, r_sig in zip(heart_signals, resp_signals):
    h_feat = extract_time_features(h_sig) + extract_freq_features(h_sig)
    r_feat = extract_time_features(r_sig) + extract_freq_features(r_sig)

    heart_features.append(h_feat)
    resp_features.append(r_feat)
```

```python
heart_feat_df = pd.DataFrame(heart_features)
heart_feat_df['label'] = heart_labels
```

```python
resp_feat_df = pd.DataFrame(resp_features)
resp_feat_df['label'] = resp_labels
```
