# **MLP**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical


# 2. Split features and labels
heart_features = heart_df.iloc[:, :-1]
resp_features = resp_df.iloc[:, :-1]
labels = heart_df.iloc[:, -1].values  # Same for both

# 3. Early fusion
fused_features = pd.concat([heart_features, resp_features], axis=1)

# 4. Normalize the features (recommended for neural nets)
scaler = StandardScaler()
X = scaler.fit_transform(fused_features)

# 5. One-hot encode labels for multi-class classification
num_classes = len(np.unique(labels))
y = to_categorical(labels, num_classes=num_classes)

# 6. Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=labels, random_state=42
)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

model = Sequential([
    Dense(128, activation='relu', input_shape=(X.shape[1],)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(num_classes, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Optional: Early stopping
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=30,
    batch_size=64,
    callbacks=[early_stop]
)
```

```python
# Accuracy
loss, acc = model.evaluate(X_test, y_test)
print(f"✅ Test Accuracy: {acc:.4f}")
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title("Model Accuracy Over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```

# **LSTM**

```python
import numpy as np
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

heart_features = heart_df.iloc[:, :-1]
resp_features = resp_df.iloc[:, :-1]
labels = heart_df.iloc[:, -1].values

# Early fusion
fused_features = pd.concat([heart_features, resp_features], axis=1)

# Scale
scaler = StandardScaler()
X = scaler.fit_transform(fused_features)

# One-hot encode labels
num_classes = len(np.unique(labels))
y = to_categorical(labels, num_classes)

# Reshape for LSTM → (samples, time_steps, features_per_step)
# We'll reshape (22,) → (11 time_steps, 2 features)
X = X.reshape((X.shape[0], 11, 2))

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=labels, random_state=42
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

model = Sequential([
    LSTM(64, input_shape=(X.shape[1], X.shape[2]), return_sequences=False),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(num_classes, activation='softmax')
])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

model = Sequential([
    LSTM(64, input_shape=(X.shape[1], X.shape[2]), return_sequences=False),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(num_classes, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train,
                    validation_data=(X_test, y_test),
                    epochs=30,
                    batch_size=64)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Get predictions from trained LSTM
y_pred_lstm = model.predict(X_test)

# Convert to class labels (for classification)
y_true = np.argmax(y_test, axis=1)
y_pred = np.argmax(y_pred_lstm, axis=1)

plt.figure(figsize=(6, 4))
plt.scatter(y_true, y_pred, alpha=0.7, s=20, color='royalblue')  # smaller, denser points
plt.plot([min(y_true), max(y_true)], [min(y_true), max(y_true)],
         color='black', linestyle='--', linewidth=2)

plt.xlabel('Actual', fontsize=12)
plt.ylabel('Predicted', fontsize=12)
plt.title('LSTM Predicted vs Actual', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
loss, acc = model.evaluate(X_test, y_test)
print(f"✅ LSTM Test Accuracy: {acc:.4f}")
```

```python
# Predict class probabilities
probs = model.predict(X_test)
# Convert probabilities to predicted class index
y_pred = np.argmax(probs, axis=1)
# If your y_test is one-hot encoded, convert to class labels
y_true = np.argmax(y_test, axis=1)
```

```python
from sklearn.metrics import classification_report, precision_score, recall_score, f1_score, accuracy_score
```

```python
# Precision, Recall, F1 Score (Weighted for multi-class)
precision = precision_score(y_true, y_pred, average='weighted')
recall = recall_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')
accuracy = accuracy_score(y_true, y_pred)

# Print all metrics
print("Weighted Precision: {:.2f}".format(precision))
print("Weighted Recall: {:.2f}".format(recall))
print("Weighted F1 Score: {:.2f}".format(f1))
print("Accuracy: {:.2f}".format(accuracy))

# Full classification report (optional, for per-class breakdown)
print(classification_report(y_true, y_pred, digits=2))
```

# **CNN + LSTM**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical

heart_features = heart_df.iloc[:, :-1]
resp_features = resp_df.iloc[:, :-1]
labels = heart_df.iloc[:, -1].values

# Early fusion
fused_features = pd.concat([heart_features, resp_features], axis=1)

# Normalize features
scaler = StandardScaler()
X = scaler.fit_transform(fused_features)

# Reshape to (samples, time_steps=11, features_per_step=2)
X = X.reshape((X.shape[0], 11, 2))

# One-hot encode labels
y = to_categorical(labels, num_classes=len(np.unique(labels)))

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=labels, random_state=42
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, LSTM, Dense, Dropout, Flatten

model = Sequential([
    Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(11, 2)),
    MaxPooling1D(pool_size=2),
    Dropout(0.3),

    LSTM(64, return_sequences=False),
    Dropout(0.3),

    Dense(64, activation='relu'),
    Dropout(0.3),

    Dense(y.shape[1], activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=30,
    batch_size=64
)
```

```python
# Predict class probabilities
probs = model.predict(X_test)
# Convert probabilities to predicted class index
y_pred = np.argmax(probs, axis=1)
# If your y_test is one-hot encoded, convert to class labels
y_true = np.argmax(y_test, axis=1)
```

```python
# Precision, Recall, F1 Score (Weighted for multi-class)
precision = precision_score(y_true, y_pred, average='weighted')
recall = recall_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')
accuracy = accuracy_score(y_true, y_pred)

# Print all metrics
print("Weighted Precision: {:.2f}".format(precision))
print("Weighted Recall: {:.2f}".format(recall))
print("Weighted F1 Score: {:.2f}".format(f1))
print("Accuracy: {:.2f}".format(accuracy))

# Full classification report (optional, for per-class breakdown)
print(classification_report(y_true, y_pred, digits=2))
```

# **BiLSTM**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Bidirectional, LSTM, Dense, Dropout

# Build the BiLSTM model
model = Sequential([
    Bidirectional(LSTM(64, return_sequences=False), input_shape=(11, 2)),
    Dropout(0.4),

    Dense(64, activation='relu'),
    Dropout(0.4),

    Dense(y_train.shape[1], activation='softmax')  # 15 classes
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.summary()
```
